```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
# Datos de ventas mensuales (en miles)
data = {
    "Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"],
    "Ventas": [25, 30, 28, 35, 40, 50, 55, 60, 70, 80, 75, 90]
}

# Crear un DataFrame de pandas
df_ventas = pd.DataFrame(data)
display(df_ventas)
```

### Análisis con NumPy

Ahora, aplicaremos algunas operaciones de NumPy para obtener estadísticas clave de las ventas.

```python
# Convertir la columna 'Ventas' a un array de NumPy para cálculos
ventas_np = np.array(df_ventas['Ventas'])

# Calcular estadísticas básicas usando NumPy
venta_total = np.sum(ventas_np)
venta_promedio = np.mean(ventas_np)
venta_maxima = np.max(ventas_np)
venta_minima = np.min(ventas_np)

# Calcular el rango de ventas (diferencia entre máxima y mínima)
rango_ventas = np.ptp(ventas_np)

# Calcular el porcentaje de crecimiento mensual
# Para evitar dividir por cero en el primer mes, empezamos desde el segundo mes
crecimiento_mensual = np.diff(ventas_np) / ventas_np[:-1] * 100

print(f"Venta Total Anual: {venta_total} mil unidades")
print(f"Venta Promedio Mensual: {venta_promedio:.2f} mil unidades")
print(f"Venta Máxima (Mes): {venta_maxima} mil unidades")
print(f"Venta Mínima (Mes): {venta_minima} mil unidades")
print(f"Rango de Ventas: {rango_ventas} mil unidades")

print("\nCrecimiento Porcentual Mensual:")
for i, mes in enumerate(df_ventas['Mes'].iloc[1:]):
    print(f"  {df_ventas['Mes'].iloc[i]} a {mes}: {crecimiento_mensual[i]:.2f}%")
```

### Visualización con Matplotlib

Finalmente, visualizaremos las ventas mensuales para identificar tendencias.

```python
plt.figure(figsize=(12, 6))
plt.plot(df_ventas['Mes'], df_ventas['Ventas'], marker='o', linestyle='-', color='skyblue', linewidth=2, markersize=8)
plt.title('Ventas Mensuales del Negocio', fontsize=16)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Ventas (en miles)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right') # Rotar etiquetas para mejor legibilidad
plt.yticks(fontsize=10)
plt.gca().yaxis.set_major_formatter(mtick.FormatStrFormatter('%d mil')) # Formato para el eje Y
plt.tight_layout() # Ajustar el diseño para evitar recortes
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
# Datos de calificaciones de estudiantes
data_calificaciones = {
    "Estudiante": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ian", "Jack"],
    "Matemáticas": [85, 92, 78, 90, 88, 76, 95, 89, 82, 94],
    "Ciencias": [90, 85, 80, 88, 92, 75, 95, 87, 81, 93],
    "Historia": [78, 82, 88, 90, 85, 80, 92, 84, 79, 91]
}

# Crear un DataFrame de pandas
df_calificaciones = pd.DataFrame(data_calificaciones)
display(df_calificaciones)
```

### Análisis con NumPy

Vamos a usar NumPy para calcular estadísticas clave de las calificaciones.

```python
# Convertir las calificaciones a arrays de NumPy para cálculos
matematicas_np = np.array(df_calificaciones['Matemáticas'])
ciencias_np = np.array(df_calificaciones['Ciencias'])
historia_np = np.array(df_calificaciones['Historia'])

# Calcular promedios por asignatura
promedio_matematicas = np.mean(matematicas_np)
promedio_ciencias = np.mean(ciencias_np)
promedio_historia = np.mean(historia_np)

# Calcular la calificación más alta y más baja en general
calificacion_maxima_general = np.max([matematicas_np, ciencias_np, historia_np])
calificacion_minima_general = np.min([matematicas_np, ciencias_np, historia_np])

# Promedio de calificaciones por estudiante (eje 1 para sumar por filas, luego dividir por el número de asignaturas)
df_calificaciones['Promedio_General'] = df_calificaciones[['Matemáticas', 'Ciencias', 'Historia']].mean(axis=1)

print(f"Promedio de Matemáticas: {promedio_matematicas:.2f}")
print(f"Promedio de Ciencias: {promedio_ciencias:.2f}")
print(f"Promedio de Historia: {promedio_historia:.2f}")
print(f"Calificación Máxima General: {calificacion_maxima_general}")
print(f"Calificación Mínima General: {calificacion_minima_general}")

print("\nPromedios individuales de los estudiantes:")
display(df_calificaciones[['Estudiante', 'Promedio_General']])
```

### Visualización con Matplotlib

Ahora, visualizaremos las calificaciones para entender mejor su distribución y rendimiento.

```python
plt.figure(figsize=(14, 7))

# Gráfico de barras de promedios por asignatura
subject_averages = df_calificaciones[['Matemáticas', 'Ciencias', 'Historia']].mean()
subject_averages.plot(kind='bar', color=['lightcoral', 'lightgreen', 'skyblue'])
plt.title('Promedio de Calificaciones por Asignatura', fontsize=16)
plt.xlabel('Asignatura', fontsize=12)
plt.ylabel('Calificación Promedio', fontsize=12)
plt.ylim(0, 100) # Rango de calificaciones de 0 a 100
plt.xticks(rotation=0) # Mantener etiquetas de asignaturas horizontales
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 7))
# Gráfico de líneas para el rendimiento de cada estudiante en las asignaturas
for subject in ['Matemáticas', 'Ciencias', 'Historia']:
    plt.plot(df_calificaciones['Estudiante'], df_calificaciones[subject], marker='o', label=subject)

plt.title('Calificaciones Individuales por Estudiante y Asignatura', fontsize=16)
plt.xlabel('Estudiante', fontsize=12)
plt.ylabel('Calificación', fontsize=12)
plt.ylim(60, 100) # Ajustar el límite y para mejor visualización
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Asignatura')
plt.tight_layout()
plt.show()
```
